```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers (1).csv')
df = pd.read_csv('Employee.csv')
df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.fillna("Unknown", inplace=True)
```

```python
df.isnull().sum()
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

```python
summary = df.describe()
summary
```

```python
summary.to_csv("summary_report.csv")
```

```python
df.isnull().sum().plot(kind='bar')
plt.title("Missing Values After Cleaning")
plt.show()
```

```python
df['SALES'].hist()
plt.title("Sales Distribution")
plt.show()
```

```python
df['PRODUCTLINE'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Product Line Distribution")
plt.ylabel('') # Hide the default 'PRODUCTLINE' y-label for pie chart clarity
plt.show()
```

```python
with open("report.txt", "w") as f:
    f.write("DATA CLEANING REPORT\n")
    f.write("--------------------\n")
    f.write(f"Rows: {df.shape[0]}\n")
    f.write(f"Columns: {df.shape[1]}\n")
    f.write("Missing values handled\n")
    f.write("Duplicates removed\n")
```
